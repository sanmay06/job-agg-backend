```python
#Importing required modules/ libraries
import csv
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd
```

```python
#url=get_url('Software Developer','Bengaluru')
#print("https://internshala.com/internships/computer-science-internship/page-{}/".format(1))

#lists to used to create dictionary of lists to create a dataframe and export to csv file
job_title_list=[]
job_url_list=[]
job_title_list = []
job_salary_list = []
job_company_name_list = []

for i in range(1,5):#loop to parse through multiple pages on the website
    response=requests.get("https://internshala.com/internships/computer-science-internship/page-{}/".format(i))
    soup=BeautifulSoup(response.text,'html.parser')
    cards=soup.find_all('div','container-fluid')
    print(len(cards))#displays number of job listings on each page just to test whether the page was parsed succesfully or not
    for card in cards:#loop to display details from elements on each page
        try:
            '''
            #one way to deal with nested divs
            job_name=soup.find('h3',{'class':'job-internship-name'})
            job_link=soup.find('a',{'class':'job-title-href'})
            print(job_name.text)
            print('https://internshala.com/'+job_link['href'])
            '''
            job_title=card.find('h3','job-internship-name').text
            print(job_title)#just checking whether the text is fetched from tag in element
            job_company_name=card.find('p','company-name').text
            print(job_company_name)
            job_salary=card.find('span','stipend').text
            print(job_salary)
            job_title_list.append(job_title)
            job_company_name_list.append(job_company_name)
            job_salary_list.append(job_salary)
            job_l=card.find('a','job-title-href')
            print('https://internshala.com/'+job_l['href']) #just checking whether the link is fetched from tag in element
            job_url_list.append('https://internshala.com/'+job_l['href'])
            
             
        except AttributeError: 
            continue
```

```python
#creating dictionary of lists
dict_jobs={'Job Title':job_title_list,'Link':job_url_list}
#creating datafram
df = pd.DataFrame(dict_jobs)
#saving as a csv file
df.to_csv('jobs.csv')
```
